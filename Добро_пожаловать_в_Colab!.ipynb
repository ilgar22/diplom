```
#       test.csv
test = pd.read_csv('/content/test.csv')
test.head(3)
```

```
# Eksik deerleri .isnull() ile tespit edelim
# ve bunlarn saysn sum() ile hesaplayalm
train.isnull().sum()
```

```
train.drop(columns = 'Cabin', axis = 1, inplace = True)
```

```
# Age (ya) kesinlikle nemli, eksik deerleri ortalama ile deitirelim.
train['Age'] = train['Age'].fillna(train['Age'].mean())
```

```
train.dropna(inplace = True)
```

```
train.isnull().sum()
```

```

# Sex (cinsiyet) deikenine pd.get_dummies() fonksiyonu ile one-hot encoding uygulayalm.
pd.get_dummies(train['Sex']).head(3)
```

```
previous = pd.read_csv('/content/train.csv')[['Sex']]
previous.head()
```

```
pd.get_dummies(previous['Sex'], dtype = int).head(3)
```

```
# lk stunu silelim, nk gereksiz.
sex = pd.get_dummies(train['Sex'], drop_first = True)
sex.head(3)
```

```

embarked = pd.get_dummies(train['Embarked'], drop_first = True)
pclass = pd.get_dummies(train['Pclass'], drop_first = True)
```

```
train = pd.concat([train, pclass, sex, embarked], axis = 1)
train.head(3)
```

```
train.drop(['PassengerId', 'Pclass', 'Name', 'Sex', 'Ticket', 'Embarked'], axis = 1, inplace = True)
train.head(3)
```

```
# StandardScaler snfn ie aktaralm
from sklearn.preprocessing import StandardScaler

# Bu snftan bir nesne oluturalm
scaler = StandardScaler()

# leklendirmek istediimiz stunlar seelim
cols_to_scale = ['Age', 'Fare']

# Verileri leklendirmek iin ortalama ve standart sapmay hesaplayalm
scaler.fit(train[cols_to_scale])

# Hesaplanan deerleri uygulayalm
train[cols_to_scale] = scaler.transform(train[cols_to_scale])

# Sonuca bakalm
train.head(3)
```

```
train.columns
```

```
train.columns = train.columns.map(str)
train.columns
```

```
# 'Survived' stunu dnda her eyi X_train'e yerletirelim
X_train = train.drop('Survived', axis = 1)

# 'Survived' stunu hedef deikenimiz (y_train) olacak
y_train = train['Survived']
```

```
# sklearn ktphanesinin linear_model modlnden lojistik regresyonu ie aktaralm
from sklearn.linear_model import LogisticRegression

# Bu snftan bir nesne oluturalm ve model deikenine atayalm
model = LogisticRegression()

# Modelimizi eitelim
model.fit(X_train, y_train)
```

```
y_pred_train = model.predict(X_train)
```

```
# Hata matrisini oluturalm
from sklearn.metrics import confusion_matrix

# Gerek ve tahmin edilen deerleri verelim
conf_matrix = confusion_matrix(y_train, y_pred_train)

# Veri erevesine dntrelim
conf_matrix_df = pd.DataFrame(conf_matrix)
conf_matrix_df
```

```
conf_matrix_labels = pd.DataFrame(conf_matrix, columns = [' ', ' '], index = [' ', ' '])
conf_matrix_labels
```

```
# sklearn'den doruluk (accuracy) metriini ie aktaralm
from sklearn.metrics import accuracy_score

# Gerek ve tahmin edilen deerleri verelim
model_accuracy = accuracy_score(y_train, y_pred_train)

#  ondalk basamaa yuvarlayalm
round(model_accuracy, 3)
```

```
# ncelikle veri setine allm isim olan X_test'i verelim
X_test = test

# Age ve Fare deikenlerindeki eksik deerleri ortalama ile dolduralm
X_test['Age'] = X_test['Age'].fillna(test['Age'].mean())
X_test['Fare'] = X_test['Fare'].fillna(test['Fare'].mean())

# Kategorik deikenlere one-hot encoding uygulayalm
sex = pd.get_dummies(X_test['Sex'], drop_first=True)
embarked = pd.get_dummies(X_test['Embarked'], drop_first=True)
pclass = pd.get_dummies(X_test['Pclass'], drop_first=True)

# Yeni stunlar orijinal dataframe'e ekleyelim
X_test = pd.concat([test, pclass, sex, embarked], axis=1)

# Artk gerekmeyen verileri silelim
X_test.drop(['PassengerId', 'Pclass', 'Name', 'Sex', 'Cabin', 'Ticket', 'Embarked'], axis=1, inplace=True)

# Sonuca bakalm
X_test.head(3)
```

```
X_test[cols_to_scale] = scaler.transform(X_test[cols_to_scale])
X_test.head(3)
```

```
X_test.columns = X_test.columns.map(str)
```

```
y_pred_test = model.predict(X_test)
```

```
# lk 10 tahmin edilen deere bakalm
y_pred_test[:10]
```

```
# Test veri setinden PassengerId stununu alalm
ids = test['PassengerId']

# Bir szlkten DataFrame oluturalm,
# ilk anahtar-deer ifti yolcu kimlii, ikinci ise test zerindeki tahmin
result = pd.DataFrame({'PassengerId': ids, 'Survived': y_pred_test})

# Sonucu grelim
result.head()
```

```
# to_csv() fonksiyonu ile yeni bir result.csv dosyas oluturalm, indeks olmadan
result.to_csv('result.csv', index=False)

# Dosya 'Oturum Deposu'nda kaydedilecek ve her ey baarl olursa u mesaj yazdracaz:
print('Dosya baaryla oturum deposuna kaydedildi!')
```
