```
import pandas as pd
```

```
# применим функцию read_csv() и посмотрим на первые три записи файла train.csv
train = pd.read_csv('/content/train.csv')
train.head(3)
```

```
# сделаем то же самое с файлом test.csv
test = pd.read_csv('/content/test.csv')
test.head(3)
```

```
train.info()
```

PassengerId — Yolcu kimlik numarası
Survived — Öldü (0) veya hayatta kaldı (1)
Pclass — Bilet sınıfı: birinci (1), ikinci (2) veya üçüncü (3)
Name — Yolcunun adı
Sex — Cinsiyet
Age — Yaş
SibSp — Gemide bulunan kardeş veya eş sayısı (siblings and spouses)
Parch — Gemide bulunan ebeveyn veya çocuk sayısı (parents and children)
Ticket — Bilet numarası
Fare — Bilet ücreti
Cabin — Kamara numarası
Embarked — Biniş limanı (C — Cherbourg; Q — Queenstown; S — Southampton)

```
# Eksik değerleri .isnull() ile tespit edelim
# ve bunların sayısını sum() ile hesaplayalım
train.isnull().sum()
```

```
train.drop(columns = 'Cabin', axis = 1, inplace = True)
```

```
# Age (yaş) kesinlikle önemli, eksik değerleri ortalama ile değiştirelim.
train['Age'] = train['Age'].fillna(train['Age'].mean())
```

```
train.dropna(inplace = True)
```

```
train.isnull().sum()
```

```

# Sex (cinsiyet) değişkenine pd.get_dummies() fonksiyonu ile one-hot encoding uygulayalım.
pd.get_dummies(train['Sex']).head(3)
```

```
previous = pd.read_csv('/content/train.csv')[['Sex']]
previous.head()
```

```
pd.get_dummies(previous['Sex'], dtype = int).head(3)
```

```
# İlk sütunu silelim, çünkü gereksiz.
sex = pd.get_dummies(train['Sex'], drop_first = True)
sex.head(3)
```

```

embarked = pd.get_dummies(train['Embarked'], drop_first = True)
pclass = pd.get_dummies(train['Pclass'], drop_first = True)
```

```
train = pd.concat([train, pclass, sex, embarked], axis = 1)
train.head(3)
```

```
train.drop(['PassengerId', 'Pclass', 'Name', 'Sex', 'Ticket', 'Embarked'], axis = 1, inplace = True)
train.head(3)
```

```
# StandardScaler sınıfını içe aktaralım
from sklearn.preprocessing import StandardScaler

# Bu sınıftan bir nesne oluşturalım
scaler = StandardScaler()

# Ölçeklendirmek istediğimiz sütunları seçelim
cols_to_scale = ['Age', 'Fare']

# Verileri ölçeklendirmek için ortalama ve standart sapmayı hesaplayalım
scaler.fit(train[cols_to_scale])

# Hesaplanan değerleri uygulayalım
train[cols_to_scale] = scaler.transform(train[cols_to_scale])

# Sonuca bakalım
train.head(3)
```

```
train.columns
```

```
train.columns = train.columns.map(str)
train.columns
```

```
# 'Survived' sütunu dışında her şeyi X_train'e yerleştirelim
X_train = train.drop('Survived', axis = 1)

# 'Survived' sütunu hedef değişkenimiz (y_train) olacak
y_train = train['Survived']
```

```
# sklearn kütüphanesinin linear_model modülünden lojistik regresyonu içe aktaralım
from sklearn.linear_model import LogisticRegression

# Bu sınıftan bir nesne oluşturalım ve model değişkenine atayalım
model = LogisticRegression()

# Modelimizi eğitelim
model.fit(X_train, y_train)
```

```
y_pred_train = model.predict(X_train)
```

```
# Hata matrisini oluşturalım
from sklearn.metrics import confusion_matrix

# Gerçek ve tahmin edilen değerleri verelim
conf_matrix = confusion_matrix(y_train, y_pred_train)

# Veri çerçevesine dönüştürelim
conf_matrix_df = pd.DataFrame(conf_matrix)
conf_matrix_df
```

```
conf_matrix_labels = pd.DataFrame(conf_matrix, columns = ['Прогноз погиб', 'Прогноз выжил'], index = ['Факт погиб', 'Факт выжил'])
conf_matrix_labels
```

```
# sklearn'den doğruluk (accuracy) metriğini içe aktaralım
from sklearn.metrics import accuracy_score

# Gerçek ve tahmin edilen değerleri verelim
model_accuracy = accuracy_score(y_train, y_pred_train)

# Üç ondalık basamağa yuvarlayalım
round(model_accuracy, 3)
```

```
# Öncelikle veri setine alışılmış isim olan X_test'i verelim
X_test = test

# Age ve Fare değişkenlerindeki eksik değerleri ortalama ile dolduralım
X_test['Age'] = X_test['Age'].fillna(test['Age'].mean())
X_test['Fare'] = X_test['Fare'].fillna(test['Fare'].mean())

# Kategorik değişkenlere one-hot encoding uygulayalım
sex = pd.get_dummies(X_test['Sex'], drop_first=True)
embarked = pd.get_dummies(X_test['Embarked'], drop_first=True)
pclass = pd.get_dummies(X_test['Pclass'], drop_first=True)

# Yeni sütunları orijinal dataframe'e ekleyelim
X_test = pd.concat([test, pclass, sex, embarked], axis=1)

# Artık gerekmeyen verileri silelim
X_test.drop(['PassengerId', 'Pclass', 'Name', 'Sex', 'Cabin', 'Ticket', 'Embarked'], axis=1, inplace=True)

# Sonuca bakalım
X_test.head(3)
```

```
X_test[cols_to_scale] = scaler.transform(X_test[cols_to_scale])
X_test.head(3)
```

```
X_test.columns = X_test.columns.map(str)
```

```
y_pred_test = model.predict(X_test)
```

```
# İlk 10 tahmin edilen değere bakalım
y_pred_test[:10]
```

```
# Test veri setinden PassengerId sütununu alalım
ids = test['PassengerId']

# Bir sözlükten DataFrame oluşturalım,
# ilk anahtar-değer çifti yolcu kimliği, ikinci ise test üzerindeki tahmin
result = pd.DataFrame({'PassengerId': ids, 'Survived': y_pred_test})

# Sonucu görelim
result.head()
```

```
# to_csv() fonksiyonu ile yeni bir result.csv dosyası oluşturalım, indeks olmadan
result.to_csv('result.csv', index=False)

# Dosya 'Oturum Deposu'nda kaydedilecek ve her şey başarılı olursa şu mesajı yazdıracağız:
print('Dosya başarıyla oturum deposuna kaydedildi!')
```
